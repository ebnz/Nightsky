```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Import Datasets (Bayer-Flamsteed-Cross Index, Hipparcos Dataset and Star Names)
===
Bayer-Flamsteed-Cross Index (IV/27A/catalog) from https://vizier.cds.unistra.fr/viz-bin/VizieR-3?-source=IV/27A/catalog

Hipparcos-Datasset (I/239/hip_main) from https://vizier.cds.unistra.fr/viz-bin/VizieR-2

Star-Names generated/obtained by Eike

```python
#Import Datasets

#Bayer-Flamsteed-Cross Index
bfi = pd.read_csv("bayer_flamsteed_cross.csv")

#Hipparcos-Dataset
hip = pd.read_csv("hipparcos.csv")

#Star Names by Eike
star_names = pd.read_csv("star_names.csv")
```

Dataset Preprocessing
===

```python
#Set Datatype to Pandas Int and Index by HIP-Column
bfi["HIP"] = bfi["HIP"].astype("Int64")
bfi = bfi.set_index("HIP")
bfi
```

```python
#Index by HIP-Column
hip = hip.set_index("HIP")
hip
```

```python
#Print NaN's
nans = bfi[bfi.index.isnull()]
nans
```

```python
#Remove NaN's
bfi = bfi[np.invert(bfi.index.isnull())]
bfi
```

```python
#Cast HIP-Column to Int. When impossible, drop row
star_names["HIP"] = pd.to_numeric(star_names["HIP"], errors="coerce")
star_names = star_names.dropna(subset=["HIP"])

#Drop old arbitrary Index-Column
star_names = star_names.drop("Unnamed: 0", axis=1).set_index("HIP")
star_names
```

```python
#Rename "IAU Name " to "name"
star_names = star_names.rename({"IAU Name ": "name"}, axis=1)
star_names = pd.DataFrame(star_names["name"])
star_names
```

Join Datasets
===
1. Join Hipparcos with Bayer-Flamsteed-Cross Index
2. Calculate Distance from Parallax
3. Calculate X, Y, Z Coordinates from Equatorial Coordinates
4. Join Result with star_names

```python
#Join Hipparcos-Catalog and Bayer-Flamsteed-Cross Index
joined = bfi.join(hip, lsuffix="_bfi", rsuffix="_hip", how="left")
joined
```

```python
#Remove Rows that have no Plx and thus no Distance can be calculated
joined = joined[np.invert(joined["Plx"].isnull())]
joined["Plx"].isnull().any()
```

```python
def calc_distance(plx):
    return 1 / (plx / 1000)
```

```python
#Calculate Distance in Parsecs for all stars
joined["distance"] = calc_distance(joined["Plx"])
joined
```

```python
#Calculate XYZ (Cartesian Coordinates) from Equatorial Coordinates
#X, Y, Z are in Parsecs
#r: radius, phi: right ascension, theta: declination
def rpt_to_x(r, phi, theta):
    return r * np.sin(np.radians(theta + 90)) * np.cos(np.radians(phi))

def rpt_to_y(r, phi, theta):
    return r * np.sin(np.radians(theta + 90)) * np.sin(np.radians(phi))

def rpt_to_z(r, phi, theta):
    return r * np.cos(np.radians(theta + 90))

joined["x"] = rpt_to_x(joined["distance"], joined["RAJ2000"], joined["DEJ2000"])
joined["y"] = rpt_to_y(joined["distance"], joined["RAJ2000"], joined["DEJ2000"])
joined["z"] = rpt_to_z(joined["distance"], joined["RAJ2000"], joined["DEJ2000"])
```

```python
#Join with star_names
joined = joined.join(star_names)
```

Export Constellations-Dataset
===

```python
#Export Dataset
joined.to_csv("constellations_dataset.csv")
#For exporting only specific columns
#joined.loc[:, ["HD", "RAJ2000", "DEJ2000", "distance", "x", "y", "z"]].to_csv("constellations_dataset_cropped_columns.csv")
```

```python
#Check that all Stars in the Connections in Bayer-Notation do also exist in the Constellations-Dataset

#For each Celestial
for starsign in conns_bayer.keys():
    
    #For each Connection in the Celestial
    for item in conns_bayer[starsign]:
        
        #Check if Start/End of Celestial exists in Constellation-Dataset
        hasstart = (joined[joined["Cst"] == starsign]["Bayer"] == item[0]).any()
        hasend = (joined[joined["Cst"] == starsign]["Bayer"] == item[1]).any()
        
        if not (hasstart and hasend):
            print(f"FAILURE: Starsign {starsign} has no star {item}!")
```

```python
#Collects the Start-HD-Index of each Connection
from_HD = []
#Collects the End-HD-Index of each Connection
to_HD = []
#Collects each Celestial in Bayer-Notation (only Constellation, no stars)
cst = []

#For each Celestial
for key in conns_bayer.keys():
    
    #Gather all Connections in Celestial in Bayer-Notation
    connections = conns_bayer[key]
    #Gather all Stars in Celestial
    set_of_cst = joined[joined["Cst"] == key]
    
    #For each Connection in Bayer-Notation, gather HD-Index of Start/End of Connection
    #If duplicates occur, use the first Element
    for item in connections:
        from_HD.append(set_of_cst[set_of_cst["Bayer"] == item[0]].iloc[0]["HD"])
        to_HD.append(set_of_cst[set_of_cst["Bayer"] == item[1]].iloc[0]["HD"])
        cst.append(key)
        
conns_HD = pd.DataFrame(data={"From": from_HD, "To": to_HD, "Cst": cst})
conns_HD
```

Translate `conns_HD` from HD-Indexing to HIP-Indexing
===
`conns_HD` contains all Connection with their Celestial in HD-Indexing

```python
#Gather Translation-Table and set to HD-Indexing
hd_to_hip = joined.loc[:, ["HD"]]
hd_to_hip = hd_to_hip.reset_index()
hd_to_hip = hd_to_hip.set_index("HD")
hd_to_hip
```

```python
#Join HIP-Indexing to each Connection and rename Columns
conns_HD_HIP = conns_HD.join(hd_to_hip, how="left", on="From", validate="many_to_one")
conns_HD_HIP = conns_HD_HIP.rename(columns={"HIP": "From_HIP", "From": "From_HD"})
conns_HD_HIP = conns_HD_HIP.join(hd_to_hip, how="left", on="To", validate="many_to_one")
conns_HD_HIP = conns_HD_HIP.rename(columns={"HIP": "To_HIP", "To": "To_HD"})
conns_HD_HIP
```
